```python
import numpy as np

def arnold_cat_map(point, n):
    A = np.array([[2, 1],
                  [1, 1]])
    return np.dot(A, point) % n

def find_period(n):
    for x in range(n):
        for y in range(n):
            initial_point = np.array([x, y])
            current_point = initial_point
            period = 0
            while True:
                current_point = arnold_cat_map(current_point, n)
                period += 1
                if np.array_equal(current_point, initial_point):
                    break
            print("Initial point ({}, {}), period: {}".format(x, y, period))

n = 1000  # Change this to the size of the grid
find_period(n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def arnold_cat_map(image, iterations):
    h, w, _ = image.shape
    mapped_image = np.empty_like(image)
    for _ in range(iterations):
        y, x = np.indices((h, w))
        mapped_yx = (2 * y + x) % h, (x + y) % w
        mapped_image[y, x] = image[mapped_yx[0], mapped_yx[1]]
        image = np.copy(mapped_image)
    return mapped_image

# Load and resize the image
image_path = 'wilson.jpg'
target_size = (1000, 1000)  # New size for the image
image = np.array(Image.open(image_path).resize(target_size))

# Apply Arnold's cat map
iterations = 1
mapped_image = arnold_cat_map(image, iterations)

# Plot original and mapped images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Mapped Image (Iterations: {iterations})')  # Update title with iteration number
plt.imshow(mapped_image)
plt.axis('off')

plt.savefig('mapped_image_1.png')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import imageio

def arnold_cat_map(image, iterations):
    h, w, _ = image.shape
    mapped_image = np.empty_like(image)
    for _ in range(iterations):
        y, x = np.indices((h, w))
        mapped_yx = (2 * y + x) % h, (x + y) % w
        mapped_image[y, x] = image[mapped_yx[0], mapped_yx[1]]
        image = np.copy(mapped_image)
        yield mapped_image  # yield each iteration's mapped image

# Load and resize the image
image_path = 'wilson.jpg'
target_size = (1000, 1000)  # New size for the image
image = np.array(Image.open(image_path).resize(target_size))

# Apply Arnold's cat map for a specific number of iterations
iterations = 750
mapped_images = arnold_cat_map(image, iterations)

# Save each iteration as an image
image_filenames = []
for i, mapped_image in enumerate(mapped_images):
    filename = f'iteration_{i}.png'
    plt.imshow(mapped_image)
    plt.title(f'Iteration {i}')  # Add frame number as title
    plt.axis('off')
    plt.savefig(filename, bbox_inches='tight', pad_inches=0)
    plt.close()
    image_filenames.append(filename)

# Create GIF from the saved images
with imageio.get_writer('arnold_cat_map_iterations.gif', mode='I', duration=0.1) as writer:
    for filename in image_filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

# Cleanup: remove temporary image files
import os
for filename in image_filenames:
    os.remove(filename)

print("GIF created successfully!")
```

```python
import numpy as np
from collections import defaultdict

def arnold_cat_map(point, n):
    A = np.array([[2, 1],
                  [1, 1]])
    return tuple((np.dot(A, point) % n).tolist())  # Convert array to tuple

def find_period(n):
    period_counts = defaultdict(int)
    visited = set()
    for x in range(n):
        for y in range(n):
            if (x, y) not in visited:
                initial_point = (x, y)
                current_point = initial_point
                period = 0
                while True:
                    current_point = arnold_cat_map(current_point, n)
                    period += 1
                    visited.add(current_point)
                    if current_point == initial_point:
                        break
                period_counts[period] += 1
    return period_counts

n = 1000  # Change this to the size of the grid
period_counts = find_period(n)
print("Periods and their frequencies:")
for period, count in period_counts.items():
    print("Period {}: {} occurrences".format(period, count))
```
