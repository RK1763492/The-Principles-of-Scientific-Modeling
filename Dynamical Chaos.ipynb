```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def arnold_cat_map(image, iterations):
    h, w, _ = image.shape
    mapped_image = np.empty_like(image)
    for _ in range(iterations):
        y, x = np.indices((h, w))
        mapped_yx = (2 * y + x) % h, (x + y) % w
        mapped_image[y, x] = image[mapped_yx[0], mapped_yx[1]]
        image = np.copy(mapped_image)
    return mapped_image

# Load and resize the image
image_path = 'wilson.jpg'
target_size = (1000, 1000)  # New size for the image
image = np.array(Image.open(image_path).resize(target_size))

# Apply Arnold's cat map
iterations = 750
mapped_image = arnold_cat_map(image, iterations)

# Plot original and mapped images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Mapped Image')
plt.imshow(mapped_image)
plt.axis('off')

plt.show()
```
