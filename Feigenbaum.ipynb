```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def bifurcation_diagram(r_values, iterations, last_iterations):
    x_values = np.empty((len(r_values), last_iterations))
    x = 0.5 * np.ones(len(r_values))

    for i in range(iterations):
        x = logistic_map(r_values, x)

    for i in range(last_iterations):
        x = logistic_map(r_values, x)
        x_values[:, i] = x

    return r_values, x_values

def plot_bifurcation_diagram(r_values, x_values):
    plt.figure(figsize=(10, 6))
    plt.title('Bifurcation Diagram')
    for i in range(x_values.shape[1]):
        plt.plot(r_values, x_values[:, i], ',b', alpha=0.25)
    plt.xlabel('r')
    plt.ylabel('x')
    plt.show()

# Parameters
start_r = 2.5
end_r = 4.0
num_points = 1000
iterations = 1000
last_iterations = 100

# Generate values for r
r_values = np.linspace(start_r, end_r, num_points)

# Generate bifurcation diagram data
r_values, x_values = bifurcation_diagram(r_values, iterations, last_iterations)

# Plot the bifurcation diagram
plot_bifurcation_diagram(r_values, x_values)

# Calculate Feigenbaum constant
delta_values = []
for i in range(1, len(r_values) - 1):
    if (x_values[i] - x_values[i - 1]) * (x_values[i] - x_values[i + 1]) > 0:
        continue
    delta = (r_values[i] - r_values[i - 1]) / (r_values[i + 1] - r_values[i])
    delta_values.append(delta)

feigenbaum_constant = np.mean(delta_values)
print("Feigenbaum constant (delta):", feigenbaum_constant)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def bifurcation_diagram(r_values, iterations, last_iterations):
    x_values = np.empty((len(r_values), last_iterations))
    x = 0.5 * np.ones(len(r_values))

    for i in range(iterations):
        x = logistic_map(r_values, x)

    for i in range(last_iterations):
        x = logistic_map(r_values, x)
        x_values[:, i] = x

    return r_values, x_values

def plot_bifurcation_diagram(r_values, x_values, xlim=None, ylim=None):
    plt.figure(figsize=(10, 6))
    plt.title('Bifurcation Diagram')
    for i in range(x_values.shape[1]):
        plt.plot(r_values, x_values[:, i], ',b', alpha=1)
    plt.xlabel('r')
    plt.ylabel('x')
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    plt.show()

# Parameters
start_r = 2.5
end_r = 4.7  # Extend the range to include 4.6
num_points = 5000  # Increase the number of points for better resolution
iterations = 5000
last_iterations = 100

# Generate values for r
r_values = np.linspace(start_r, end_r, num_points)

# Generate bifurcation diagram data
r_values, x_values = bifurcation_diagram(r_values, iterations, last_iterations)

# Plot the entire bifurcation diagram
plot_bifurcation_diagram(r_values, x_values, xlim=(start_r, end_r))

# Zoom into a specific region
xlim = (3.7, 3.9)  # Adjust these values to zoom into different regions
ylim = (0, 1)
plot_bifurcation_diagram(r_values, x_values, xlim=xlim, ylim=ylim)

```
