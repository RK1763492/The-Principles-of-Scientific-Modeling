```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

def frequency_to_note(frequency):
    """
    Convert frequency to the closest musical note.
    """
    A4_freq = 440  # Frequency of A4 (440 Hz)
    note_names = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
    semitone_ratio = 2 ** (1/12)  # Ratio between adjacent semitones

    # Ensure frequency is positive
    frequency = abs(frequency)

    # Calculate the distance in semitones from A4
    distance_from_A4 = 12 * np.log2(frequency / A4_freq)

    # Round to the nearest integer number of semitones
    semitones = round(distance_from_A4)

    # Map the number of semitones to a note name
    note_index = (semitones + 9) % 12  # Shift by 9 to start at A
    note_name = note_names[note_index]

    # Calculate the octave
    octave = 4 + (semitones // 12)

    return f"{note_name}{octave}", frequency

# Load the WAV file
sample_rate, audio_data = wavfile.read(r"C:\Users\rkesh\OneDrive - Villanova University\3rd Year\Semester 2\Modeling\Assignment 10\untitled.wav")

# Define the time axis
time = np.arange(len(audio_data)) / sample_rate

# Define parameters for segmenting the audio data
segment_duration = 0.5  # Duration of each segment in seconds
overlap = 0.25  # Overlap between segments in seconds

# Calculate the number of samples per segment
segment_samples = int(segment_duration * sample_rate)
overlap_samples = int(overlap * sample_rate)

# Compute the Fourier transform of the entire audio file
fourier_transform = np.fft.fft(audio_data)
freqs = np.fft.fftfreq(len(audio_data), d=1/sample_rate)

# Plot the Fourier transform
plt.figure(figsize=(10, 4))
plt.plot(time, np.abs(fourier_transform))
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of the Audio File')
plt.show()

# Iterate over segments
for i in range(0, len(audio_data) - segment_samples, segment_samples - overlap_samples):
    segment = audio_data[i:i + segment_samples]

    # Apply windowing (optional)
    window = np.hamming(len(segment))
    segment_windowed = segment * window

    # Compute the Fourier transform
    fourier_transform = np.fft.fft(segment_windowed)
    freqs = np.fft.fftfreq(len(segment), d=1/sample_rate)

    # Find the dominant frequency (fundamental pitch)
    dominant_frequency_index = np.argmax(np.abs(fourier_transform))
    dominant_frequency = freqs[dominant_frequency_index]

    # Convert the dominant frequency to the closest musical note
    note, freq_of_note = frequency_to_note(abs(dominant_frequency))

    print(f"Segment {i//segment_samples + 1}: Approximate Note: {note}")
    print(f"Frequency of the Note: {freq_of_note} Hz")
```
