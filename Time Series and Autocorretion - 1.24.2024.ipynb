```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating the array of white noise
x = np.random.normal(loc = 1, scale = 0.1, size = 1000000)
```

```python
x
```

```python
# plotting the histograms with varying bin sizes
plt.figure(figsize = [10, 7])

plt.hist(x, bins = 10, alpha = 0.3, label = '10 bins')
plt.hist(x, bins = 20, alpha = 0.5, label = '20 bins')
plt.hist(x, bins = 50, alpha = 0.7, label = '50 bins')
plt.hist(x, bins = 100, alpha = 0.85, label = '100 bins')
plt.hist(x, bins = 1000, label = '1000 bins')

plt.legend()
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Frequency of Number in the White Noise Timeseries')
```

```python
# outliers
mean = np.mean(x)
std = np.std(x)

threshold = 1

while threshold < 6:
    outliers = x[np.abs(x - mean) > (threshold * std)]
    size = np.size(outliers)
    percent = (size/np.size(x)) * 100
    print('The', threshold, '-sigma outliers:', outliers)
    print('The number of outliers:', size)
    print('The percent of outliers:', percent)
    threshold += 1
```

```python
# cumulative timeseries
cum = np.cumsum(x)
cum
```

```python
# array of 1 to 1e6
y = np.linspace(1, 1000001, 1000000)
```

```python
# least squares line
b, a = np.polyfit(cum, y, deg = 1)
```

```python
# plotting the cumulative timeseries and least squares line
plt.figure()
plt.scatter(cum, y, color = 'purple')
plt.plot(cum, a + b * cum, color = 'cyan')
plt.xlabel('Time')
plt.ylabel('Cumulative Value')
plt.title('Cumulative Time Series')
print('The slope:', b)
print('The intercept:', a)
```

```python
# autocorrelation of time series

import statsmodels.graphics.tsaplots as sm

sm.plot_acf(x, alpha = None);
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
```

```python
# graphing the original time series

plt.scatter(y, x)
plt.ylim(0, 10)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series')
```

```python
# generating two more time series

ts1 = x = np.random.normal(loc = 1, scale = 0.1, size = 1000)
ts2 = x = np.random.normal(loc = 1, scale = 0.1, size = 1000)
```

```python
# chi square test

from scipy import stats

stats.chisquare(ts1)
```

```python
stats.chisquare(ts2)
```

```python
# student's t-test

stats.ttest_ind(ts1, ts2)
```

```python
# kolmogorov-smirnov test

stats.ks_2samp(ts1, ts2)
```

```python
# Generate a sine wave
Fs = 1000  # Sampling frequency
T = 1 / Fs  # Sampling interval
t = np.arange(0, 1, T)  # Time vector
f = 50  # Frequency of the sine wave
A = 1  # Amplitude of the sine wave
sinewave = A * np.sin(2 * np.pi * f * t)

# Plot the original sine wave
plt.subplot(2, 1, 1)
plt.plot(t, sinewave)
plt.title('Original Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
sm.plot_acf(sinewave, alpha = None);
plt.xlim(0,40)
```
